# XOR data

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import utils
```

```python
scenarios: tuple = ((10, True), (10, False), (20, False), (30, False))
```

```python
def build_datasets(scenarios):
    for tup in scenarios:
        X, Y, _ = utils.generate_distance(*tup)
        df = pd.DataFrame(X)
        df['Y'] = Y
        pattern: str = 'short' if tup[1] else 'long'
        df.to_csv(f'data/xor{tup[0]}_{pattern}.csv')
```

## XOR 2D

```python
N: int = 2000
X = np.random.random((N, 2)) * 2 - 1
Y = [0 if (utils.encode_to_bits(x[0]) ^ utils.encode_to_bits(x[1])) == 1 else 1 for x in X]

plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolor='k')
# plt.savefig('out/plots/xor2d.png', dpi=600, transparent=True)
plt.show()
```

## XOR 3D

```python
X_ = np.random.random((N, 3)) * 2 - 1
Y_ = [utils.get_label(x, 0, 1, 2) for x in X_]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_[:, 0], X_[:, 1], X_[:, 2], c=Y_)
# plt.savefig('out/plots/xor3d.png', dpi=600, transparent=True)
plt.show()
```
